### Preparing the data

#### iterating of files

```python
#going through files in a folder
from pathlib import Path

DATA_PATH = Path.cwd() / "data_test"

markdown_data = []
for file in DATA_PATH.glob(pattern='*.md'):
    with open(file, "r") as f:
        text_data = f.read()
    markdown_data.append(text_data)
    video_title = file.stem
    print(video_title)
print(len(markdown_data))
```

#### iterating within files

```python
#going through parts of a file
example_file = DATA_PATH / "Course structure for Azure two weeks course.md"
list_content = []
with open(example_file, "r") as f:
    content = f.read()
list_content = content.split('\n\n')
print(len(list_content))
```
