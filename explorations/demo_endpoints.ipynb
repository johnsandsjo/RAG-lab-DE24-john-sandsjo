# Demo endpoints

```python
import requests
import os
from dotenv import load_dotenv

load_dotenv

base_url = f"https://rag-youtube.azurewebsites.net/"
```

## Main endpoint (rag/query)

```python
endpoint = f"rag/query?code={os.getenv('FUNCTION_APP_API')}"

main_response = requests.post(url=base_url+endpoint, json={"prompt" : "Video about sql"})
main_response.json()
```

## History endpoint

```python
endpoint = f"rag/history?code={os.getenv('FUNCTION_APP_API')}"

history_response = requests.get(url=base_url+endpoint)
history_response.json()
```

## Description generation endpoint

```python
endpoint = f"rag/description-gen?code={os.getenv('FUNCTION_APP_API')}"

desc_response = requests.post(url=base_url+endpoint, json={"prompt" : "Video about sql"})
desc_response.json()
```

## Keywords generator endpoint

```python
endpoint = f"rag/keyword-gen?code={os.getenv('FUNCTION_APP_API')}"

keyword_response = requests.post(url=base_url+endpoint, json={"prompt" : "Video about sql"})
keyword_response.json()
```
